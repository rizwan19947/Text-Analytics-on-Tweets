```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.cluster import KMeans
from twython import Twython
from textblob import TextBlob
import pandas as pd
import nltk
```

# Extraction

```python
consumerkey = ''
consumersecret = ''

accesstoken = ''
accesstokensecret = ''

tweets = Twython(consumerkey, consumersecret, accesstoken, accesstokensecret)
```

```python
timeline1 = tweets.get_user_timeline(screen_name = "Palestine_UN", count = 200)
timeline1 = pd.DataFrame(timeline1)
timeline1
```

```python
timeline2 = tweets.get_user_timeline(screen_name = "AJGaza", count = 200)
timeline2 = pd.DataFrame(timeline2)
timeline2
```

```python
timeline3 = tweets.get_user_timeline(screen_name = "EASPORTSFIFA", count = 200)
timeline3 = pd.DataFrame(timeline3)
timeline3
```

```python
timeline4 = tweets.get_user_timeline(screen_name = "Ubisoft", count = 200)
timeline4 = pd.DataFrame(timeline4)
timeline4
```

# Sentiment Analysis

```python
def compute_sentiment(timeline1):
    blob = TextBlob(timeline1)
    return blob.sentiment_assessments.polarity
```

```python
timeline1['Polarity'] = timeline1['text'].apply(compute_sentiment)
timeline1[['text','Polarity']]
```

```python
def compute_sentiment(timeline2):
    blob = TextBlob(timeline2)
    return blob.sentiment_assessments.polarity

timeline2['Polarity'] = timeline2['text'].apply(compute_sentiment)
timeline2[['text','Polarity']]
```

```python
def compute_sentiment(timeline3):
    blob = TextBlob(timeline3)
    return blob.sentiment_assessments.polarity

timeline3['Polarity'] = timeline3['text'].apply(compute_sentiment)
timeline3[['text','Polarity']]
```

```python
def compute_sentiment(timeline4):
    blob = TextBlob(timeline4)
    return blob.sentiment_assessments.polarity

timeline4['Polarity'] = timeline4['text'].apply(compute_sentiment)
timeline4[['text','Polarity']]
```

# Pre-Processing

```python
#Converting dict data type into string
text1 = str(timeline1)
text2 = str(timeline2)
text3 = str(timeline3)
text4 = str(timeline4)

#Converting all letters to lower case

text1 = text1.lower()
text2 = text2.lower()
text3 = text3.lower()
text4 = text4.lower()

#Removing Punctation

puncts = ".?!,';:-()[]"

for sym in puncts:
    text1 = text1.replace(sym, ' ')

for sym in puncts:
    text2 = text2.replace(sym, ' ')

for sym in puncts:
    text3 = text3.replace(sym, ' ')

for sym in puncts:
    text4 = text4.replace(sym, ' ')

```

# Bag of Words Twitter extraction for 'Palestine_UN'

```python
cnt_vectorizer = CountVectorizer(stop_words='english')
cnt_vectorizer.fit(timeline1)
print(cnt_vectorizer.vocabulary_)
print('')
print('')
print('')

print(cnt_vectorizer.get_feature_names())
```

```python
vec1 = cnt_vectorizer.transform(timeline1).toarray()
print(vec1.shape)
vec1
```

```python
km = KMeans(n_clusters=10) #Clusters are a rough approximation of K = sq.root of(n/2)
km.fit(vec1)
```

```python
labels = km.labels_
labels
```
